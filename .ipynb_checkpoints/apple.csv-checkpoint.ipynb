```python
df = pd.read_csv('https://raw.githubusercontent.com/noahwaterfieldprice/python_primer/master/ch_6/Apple.csv', index_col='Tarih',parse_dates=True) 
df = df.sort_index()
print(df.info())
```

```python
df.loc['2013', 'Close'].mean()
print (df.loc['2013', 'Close'].mean())
```
